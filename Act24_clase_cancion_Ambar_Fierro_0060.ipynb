## clase cancion

###
    - Atributos: titulo, artista, album, duracion
    - Funciones: reproducir(), agregarPlaylist(), compartir()

![texto del vínculo](https://raw.githubusercontent.com/a25308051280060-pixel/-Act24-clase-cancion-Ambar-Fierro-0060/refs/heads/main/clasecancion2.png)

```python
class Cancion:
    def __init__(self, titulo, artista, album, duracion):
        self.titulo = titulo
        self.artista = artista
        self.album = album
        self.duracion = duracion
        self.en_reproduccion = False

    def reproducir(self):
        self.en_reproduccion = True
        print(f"Reproduciendo ahora: '{self.titulo}' de {self.artista}...")

    def agregarPlaylist(self, nombre_playlist):
        print(f"La canción '{self.titulo}' ha sido agregada a la playlist: {nombre_playlist}.")

    def compartir(self):
        print(f"Enlace de compartir generado para: {self.titulo} - {self.artista}.")

    def mostrar_info(self):
        print(f"Track: {self.titulo} | Artista: {self.artista} | Álbum: {self.album} | Duración: {self.duracion}")

# --- Instanciación ---
cancion1 = Cancion("Blinding Lights", "The Weeknd", "After Hours", "3:20")
cancion2 = Cancion("Bohemian Rhapsody", "Queen", "A Night at the Opera", "5:55")

# --- Llamadas a las funciones para cada objeto ---
print("--- Info Canción 1 ---")
cancion1.mostrar_info()
cancion1.reproducir()
cancion1.agregarPlaylist("Favoritos 2024")
cancion1.compartir()

print("\n--- Info Canción 2 ---")
cancion2.mostrar_info()
cancion2.reproducir()
cancion2.agregarPlaylist("Clásicos del Rock")
cancion2.compartir()

print("\n" + "-"*30)
print("creador de software por Ambar Fierro NL = 19")
print("programa terminado")
```
